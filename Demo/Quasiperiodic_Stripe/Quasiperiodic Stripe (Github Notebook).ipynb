# Quasiperiodic Stripe

In this notebook we will generate a Quasiperiodic Stripe generated with two Star Vectors, fixing the integer number associated to the orthogonal line to one Star Vector and varying the integer number associated with the other Star Vector.

First, let's load the Julia files with the functions that we will use in this notebook

```julia
include("Operaciones_Basicas.jl")
include("Red_Dual.jl")
include("Separacion_Arreglos_X_Y.jl")
include("Franja_Cuasiperiodica.jl")
```

We will use the Plots package to visualize the Quasiperiodic Stripe, so we will load that

```julia
using Plots
```

### Obtain the vertices for the Quasiperiodic Stripes

#### Symmetry of a Pentagon

Create the information of the star vectors required by the GDM algorithm, which determine the symmetry of the quasiperiodic lattice. Also, generate the array that contains the alpha constants.

```julia
NSides = 5; #Symmetry of the quasiperiodic lattice

Star_Vectors = [[BigFloat(1),0]]; #Array wich will contain the Star Vectors

for i in 1:(NSides-1)
    push!(Star_Vectors, [cos((2*i)*pi/NSides), sin((2*i)*pi/NSides)]); #Complete the Star_Vector Array
end

Alphas_Array = fill(0.2, NSides); #Array of the alphas constant
```

Let's do the particular case of $J = 1$, this is fixing the Star Vector collinear to the X axis.

For the symmetry considered, any other case is exactly the same as this except for a rotation.

```julia
#Integer that generate the interval [-N,N] in which the integer of the orthogonal line to the Star Vector Ek exist.
#A greater value gives a larger Stripe.
N_Interval_K = 8;

J = 1; #Index of the Star Vector, in this case, the first one. The Stripe will be "orthogonal" to this vector.
Nj = 0; #Integer number for the orthogonal line to the Star Vector Ej. This parameter displace the Stripe along the X axis.
Vertices_Stripe,Info_Bands = franjas_Cuasiperiodicas(J, Nj, N_Interval_K, Star_Vectors, Alphas_Array);

X_Coordinates_Stripe, Y_Coordinates_Stripe = separacion_Arreglo_de_Arreglos_2D(Vertices_Stripe);
```

```julia
#Let's visualize the stripe
plot()
for i in 1:4:length(X_Coordinates_Stripe)
    plot!([X_Coordinates_Stripe[i],X_Coordinates_Stripe[i+1],X_Coordinates_Stripe[i+2],X_Coordinates_Stripe[i+3],X_Coordinates_Stripe[i]],[Y_Coordinates_Stripe[i],Y_Coordinates_Stripe[i+1],Y_Coordinates_Stripe[i+2],Y_Coordinates_Stripe[i+3],Y_Coordinates_Stripe[i]], markersize = 0.2, key = false, aspect_ratio=:equal, grid = false, color =:black)
end
plot!(xlimits = [-35,35], ylimits = [-35,35], markersize = 0.4, key = false, aspect_ratio=:equal, grid = false, color =:black)
```

Note that due to the nature of the algorithm, the strip is made up of a main cluster and isolated polygons.

Let's do the particular case of $J = 2$.

```julia
#Integer that generate the interval [-N,N] in which the integer of the orthogonal line to the Star Vector Ek exist.
#A greater value gives a larger Stripe.
N_Interval_K = 8;

J = 2; #Index of the Star Vector, in this case, the first one. The Stripe will be "orthogonal" to this vector.
Nj = 0; #Integer number for the orthogonal line to the Star Vector Ej. This parameter displace the Stripe along the X axis.
Vertices_Stripe,Info_Bands = franjas_Cuasiperiodicas(J, Nj, N_Interval_K, Star_Vectors, Alphas_Array);

X_Coordinates_Stripe, Y_Coordinates_Stripe = separacion_Arreglo_de_Arreglos_2D(Vertices_Stripe);
```
